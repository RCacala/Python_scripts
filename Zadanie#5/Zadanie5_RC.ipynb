## Zadanie #1

```python
import os

def find_pdf_size(top):
    total_size = 0
    
    for root, dirs, files in os.walk(top):
        dir_sum = sum(
            os.path.getsize(os.path.join(root, name))
            for name in files if name.lower().endswith(".pdf")
        )
        total_size += dir_sum
    
    return total_size
```

```python
find_pdf_size(".")
```

## Zadanie #2

```python
from datetime import datetime, timedelta

def print_working_days(date1, date2):
    start = datetime.strptime(date1, "%Y-%m-%d").date()
    end = datetime.strptime(date2, "%Y-%m-%d").date()
    
    current = start
    while current < end:
        if current.weekday() <= 4:
            print(current)
        current += timedelta(days=1)
```

```python
print_working_days("2026-05-01", "2026-05-31")
```

## Zadanie #3

```python
import random

def random_walk(start):
    x = start
    while True:
        yield x
        step = random.choice([-1, 1])
        x += step
```

```python
count = 0

while count < 10:
    walker = random_walk(0)
    pos = None
    for _ in range(101):
        pos = next(walker)
    print("Final position:", pos)
    count += 1
```
