```python
word = input("Podaj dowolne słowo: ")

print("+---" * len(word) + "+")
for letter in word:
    print("|{:^3}".format(letter), end="")
print("|")
print("+---" * len(word) + "+")
```

## Zadanie #2

```python
for x in range(1, 41):
    if x == 13:
        continue

    if x % 5 == 0 and x % 7 == 0:
        print(f"{x} is divided by 5 and 7")
    elif x % 5 == 0:
        print(f"{x} is divided by 5")
    elif x % 7 == 0:
        print(f"{x} is divided by 7")
    else:
        print(f"{x} is not important")
```

```python
x = 1

while x <= 40:
    if x == 13:
        x += 1
        continue

    if x % 5 == 0 and x % 7 == 0:
        print(f"{x} is divided by 5 and 7")
    elif x % 5 == 0:
        print(f"{x} is divided by 5")
    elif x % 7 == 0:
        print(f"{x} is divided by 7")
    else:
        print(f"{x} is not important")

    x += 1
```
